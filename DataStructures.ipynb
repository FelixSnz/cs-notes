# **What are data structures?**
---

A data structure is a specialized format for organizing, storing, and retrieving data in a computer so that it can be used efficiently.

It is also used for processing, retrieving, and storing data. There are different basic and advanced types of data structures that are used in almost every program or software system that has been developed.

# **Data Types**


---



*   **Primitive**

    examples: int, char, float, double, and pointer (can hold a single value)
    
*   **Non-primitive**

    Are more complex data types, such as arrays, linked lists, stacks, queues,trees, graphs, etc. They are constructed from primitive data types and provide more complex methods for storing and manipulating data, such as searching, sorting, and insertion. Non-primitive data structures are implemented using classes or data structures in a programming language and are not a part of the language itself.

    image.png






# **Linear Data Structure**


---


A linear data structure is a type of data structure where the elements are stored in a linear or sequential manner. In other words, elements are stored one after the other and there is a clear starting and ending point to the data.



*   **Static Data Structures:**
  
    These are data structures where the memory allocation is fixed and the size of the data structure cannot be changed during the execution of the program. Arrays are an example of a static data structure.



*   **Dynamic Data Structures:**

    These are data structures where the memory allocation is flexible and can be changed during the execution of the program. Linked lists and stacks are examples of dynamic data structures.



# **Non-linear data structure**


---

are a type of data structure where the elements are stored in a non-sequential or non-linear manner. In other words, there is no clear starting and ending point to the data and the elements are stored in a way that allows for more complex relationships between them. The following are examples of non-linear data structures:

Trees: a hierarchical data structure where each node has a parent and zero or more children

Graphs: a non-linear data structure consisting of nodes and edges

# **Abstract Data Type (ADT)**


---

An Abstract Data Type (ADT) is a high-level description of a data structure that defines the behavior of a collection of data, but not the implementation details.

An ADT defines the operations that can be performed on the data and the relationships between the data, but it doesn't specify how the data is actually stored in memory.

### Singly Linked List

```python
class Node:
    def __init__(self, data, next):
        self.data = data
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None
    
    @property
    def length(self):
        count = 0
        current_node = self.head
        while current_node:
            count += 1
            current_node = current_node.next
        return count
  
    def insert_at_begining(self, data):
        self.head = Node(data, self.head)

    def insert_at_end(self, data):
        if self.head is None:
            self.head = Node(data, None)
            return
        current_node = self.head
        while current_node.next:
            current_node = current_node.next
        current_node.next = Node(data, None)
    
    def insert_values(self, values):
        self.head = None
        for value in values:
            self.insert_at_end(value)
    
    def remove_at(self, index):
        if index<0 or index>=self.length:
            raise Exception("invalid index")
        if index == 0:
            self.head = self.head.next
            return
        count = 0
        current_node = self.head

        while current_node:
            if count == index - 1:
                current_node.next = current_node.next.next
                break
            count += 1
            current_node = current_node.next
    
    def insert_at(self, index, data):
        if index<0 or index>=self.length:
            raise Exception("invalid index")
        
        if index == 0:
            self.insert_at_begining(self, data)
        
        count = 0
        current_node = self.head

        while current_node:
            if count == index -1:
                current_node.next = Node(data, current_node.next)
                break
            count +=1 
            current_node = current_node.next

    def insert_after_value(self, data_after, data_to_insert):
        # Search for first occurance of data_after value in linked list
        # Now insert data_to_insert after data_after node

        if self.head is None:
            return
        
        current_node = self.head

        while current_node:
            if current_node.data == data_after:
                current_node.next = Node(data_to_insert, current_node.next)
                break
            current_node = current_node.next

    def remove_by_value(self, data):

        if self.head is None:
            return

        if self.head.data == data:
            self.head = self.head.next
            return

        current_node = self.head

        while current_node.next:
            if current_node.next.data == data:
                current_node.next = current_node.next.next
                break

            current_node = current_node.next

  
    def traverse(self):
        if self.head is None:
            print("linked list is empty")
            return
        current_node = self.head
        something = ''
        while current_node:
            something += f"{current_node.data}-->"
            current_node = current_node.next
        print(something)
  

if __name__ == "__main__":
    ll = LinkedList()
    ll.insert_values(["banana","mango","grapes","orange"])
    ll.traverse()
    ll.insert_after_value("banana","apple") # insert apple after mango
    ll.traverse()
    ll.remove_by_value("orange") # remove orange from linked list
    ll.traverse()
    ll.remove_by_value("figs")
    ll.traverse()
    ll.remove_by_value("banana")
    ll.remove_by_value("mango")
    ll.remove_by_value("apple")
    ll.remove_by_value("grapes")

    print(ll.length)

    ll.traverse()
```

### Doubly Linked List

```python
class Node:
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev

class DoublyLinkedList:
    def __init__(self):
        self.head = None
    
    @property
    def length(self):
        count = 0
        current_node = self.head
        while current_node:
            count += 1
            current_node = current_node.next
        return count
  
    def insert_at_begining(self, data):
        self.head = Node(data, self.head)

    def insert_at_end(self, data):
        if self.head is None:
            self.head = Node(data, None)
            return
        current_node = self.head
        while current_node.next:
            current_node = current_node.next
        current_node.next = Node(data, None, current_node)
    
    def insert_values(self, values):
        self.head = None
        for value in values:
            self.insert_at_end(value)
    
    def remove_at(self, index):
        if index<0 or index>=self.length:
            raise Exception("invalid index")
        if index == 0:
            self.head = self.head.next
            return
        count = 0
        current_node = self.head

        while current_node:
            if count == index - 1:
                current_node.next = current_node.next.next
                break
            count += 1
            current_node = current_node.next
    
    def insert_at(self, index, data):
        if index<0 or index>=self.length:
            raise Exception("invalid index")
        
        if index == 0:
            self.insert_at_begining(self, data)
        
        count = 0
        current_node = self.head

        while current_node:
            if count == index -1:
                current_node.next = Node(data, current_node.next, current_node)
                break
            count +=1 
            current_node = current_node.next

    def insert_after_value(self, data_after, data_to_insert):
        # Search for first occurance of data_after value in linked list
        # Now insert data_to_insert after data_after node

        if self.head is None:
            return
        
        current_node = self.head

        while current_node:
            if current_node.data == data_after:
                #current_node.next = Node(data_to_insert, current_node.next, current_node)
                node = Node(data_to_insert, current_node.next, current_node)
                if node.next:
                    node.next.prev = node
                current_node.next = node
                break
            current_node = current_node.next

    def remove_by_value(self, data):

        if self.head is None:
            return

        if self.head.data == data:
            self.head = self.head.next
            return

        current_node = self.head

        while current_node.next:
            if current_node.next.data == data:
                current_node.next = current_node.next.next
                break

            current_node = current_node.next

  
    def traverse_forward(self):
        if self.head is None:
            print("linked list is empty")
            return
        current_node = self.head
        something = ''
        while current_node:
            something += f"{current_node.data}-->"
            current_node = current_node.next
        print(something)
    
    def traverse_backward(self):
        if self.head is None:
            print("linked list is empty")
            return
        current_node = self.head
        traversed_str = ''

        while current_node.next:
            current_node = current_node.next

        while current_node.prev:
            traversed_str += f"<--{current_node.data}"
            current_node = current_node.prev
        traversed_str += f"<--{current_node.data}"
        print(traversed_str)
  

if __name__ == "__main__":
    dll = LinkedList()
    dll.insert_values(["banana","mango","grapes","orange"])
    dll.traverse_forward()
    dll.insert_after_value("mango", "apple")
    dll.traverse_forward()
    dll.traverse_backward()


    print(ll.length)
```
